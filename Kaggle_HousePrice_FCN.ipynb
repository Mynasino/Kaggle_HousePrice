```python
import random
batch_size = 128
num_examples = 1460
num_folds = 5
size_fold = (int)(num_examples / num_folds)
def k_fold_data_iter(k):
    if k == 0:
        idx = list(range(size_fold, num_examples))
    if k + 1 == num_folds:
        idx = list(range(0, k * size_fold))
    if k != 0 and k + 1 != num_folds:
        idx = list(range(0, k * size_fold)) + list(range((k + 1) * size_fold, num_examples))
    random.shuffle(idx)
    for i in range(0, (num_folds - 1) * size_fold, batch_size):
        j = np.array(idx[i:min(i + batch_size, (num_folds - 1) * size_fold)])
        yield np.take(X_data, j, axis=0), np.take(y_data, j).reshape([min(i + batch_size, (num_folds - 1) * size_fold) - i, 1])
        #tensory
        #np.take(X_data, j)axis=0
        #np.take(y_data, j).shape(min(i + batch_size, num_examples) - i,)feed(?,1)y_,batch_size
```

```python
def xavier_init(fan_in,fan_out,constant = 1):
    low = -constant * np.sqrt(6.0/(fan_in+fan_out))
    high = constant * np.sqrt(6.0/(fan_in+fan_out))
    return tf.random_uniform((fan_in,fan_out),minval = low,maxval = high,dtype = tf.float32)
```

```python
in_units = 208
h1_units = 128
h2_units = 64
h3_units = 32
h4_units = 8
W1 = tf.Variable(xavier_init(in_units,h1_units))
b1 = tf.Variable(tf.constant(0.1, shape=[h1_units]))
W2 = tf.Variable(xavier_init(h1_units,h2_units))
b2 = tf.Variable(tf.constant(0.1, shape=[h2_units]))
W3 = tf.Variable(xavier_init(h2_units,h3_units))
b3 = tf.Variable(tf.constant(0.1, shape=[h3_units]))
W4 = tf.Variable(xavier_init(h3_units,h4_units))
b4 = tf.Variable(tf.constant(0.1, shape=[h4_units]))
W5 = tf.Variable(xavier_init(h4_units,1))
b5 = tf.Variable(tf.constant(0.1, shape=[1]))
```

```python
x = tf.placeholder(tf.float32,[None, in_units])
keep_prob = tf.placeholder(tf.float32)
```

```python
hidden1 = tf.nn.relu(tf.matmul(x,W1) + b1)
hidden1_drop = tf.nn.dropout(hidden1, keep_prob)
hidden2 = tf.nn.relu(tf.matmul(hidden1_drop,W2) + b2)
hidden2_drop = tf.nn.dropout(hidden2, keep_prob)
hidden3 = tf.nn.relu(tf.matmul(hidden2_drop,W3) + b3)
hidden3_drop = tf.nn.dropout(hidden3, keep_prob)
hidden4 = tf.nn.relu(tf.matmul(hidden3_drop,W4) + b4)
hidden4_drop = tf.nn.dropout(hidden4, keep_prob)
y = tf.matmul(hidden4_drop, W5) + b5
```

```python
y_ = tf.placeholder(tf.float32, [None, 1])
loss = tf.reduce_mean(tf.square(y - y_))
train_step = tf.train.AdagradOptimizer(0.03).minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
train_loss = []
text_loss = []

k = 0
for i in range(4000):
    for batch_x, batch_y in k_fold_data_iter(k):
        sess.run(train_step, feed_dict={x: batch_x, y_: batch_y, keep_prob: 0.99}) #feed y_ not y
    if i % 100 == 0:
        train_id = list(range(0, k * size_fold)) + list(range((k + 1) * size_fold, num_examples))
        text_id = list(range(k * size_fold, (k + 1) * size_fold))
        train_loss.append(sess.run(loss, feed_dict={x:np.take(X_data,train_id,axis=0), y_:np.take(y_data,train_id).reshape([(num_folds - 1) * size_fold,1]), keep_prob:1}))
        text_loss.append(sess.run(loss, feed_dict={x:np.take(X_data,text_id,axis=0), y_:np.take(y_data,text_id).reshape([size_fold,1]), keep_prob:1}))
        print('epoch %d' % i, 'train loss', train_loss[-1],  'text loss', text_loss[-1])
    if (train_loss[-1] * 4 + text_loss[-1]) / 5 < 0.0005:
        break
    k = (k + 1) % num_folds
```
